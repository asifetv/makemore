```python
# Initialize the training set, labels and parameters
# Implement forward pass, backward pass and updates, define loss function
# Implement the prediction loop and Generator
# 
# Read the file, create dictionary for conversion from character to index and back
# read the file and create pairs of characters from which create x_data, y_data
words = open('names.txt', 'r').read ().splitlines()
stoi = {ch:i+1 for i, ch in enumerate(sorted(set("".join(words))))}
stoi['.'] = 0 
itos = {i:ch for ch, i in stoi.items()}
#print(itos)
x_data = []
y_data = []
#print(len(words))
for w in words:
    name = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(name, name[1:]):
        #print(f'{ch1}{ch2}')
        x_data.append(stoi[ch1])
        y_data.append(stoi[ch2])

import torch
import torch.nn.functional as F

xs = torch.tensor(x_data)
ys = torch.tensor(y_data)
num = xs.nelement()

g = torch.Generator().manual_seed(2147483647)
#print(xenc.shape, y.shape, xenc, y)

#randomly initialize weights
W = torch.randn((27, 27), generator=g, dtype=torch.float32, requires_grad=True)

#Gradient Descent
for _ in range(500):
    #Forward pass
    xenc =  F.one_hot(xs, num_classes=27).float()
    logits = xenc @ W
    counts = logits.exp()
    probs = counts/counts.sum(1, keepdim=True)
    loss = -probs[torch.arange(num), ys].log().mean() + 0.01*(W**2).mean()
    print(loss.item())

    #backward pass & update
    W.grad=None
    loss.backward()

    #update
    W.data += -50 * W.grad

#print(W.shape)
```

```python
#Predict & Generate
for i in range(5):
    idx = 0
    name = []
    while True:
        xenc = F.one_hot(torch.tensor([idx]), num_classes=27).float()
        #print(xenc)
        logits = xenc @ W
        counts = logits.exp()
        p = counts/counts.sum(1, keepdim=True)
        #print(p)
        idx = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        #print(itos[idx])
        name.append(itos[idx])
        if idx == 0:
            break
    print("".join(name))
```
